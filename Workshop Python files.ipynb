# Safe dict reading

define a function `safe_dict(d, k)` that takes in a python dict `d` and a key `k` and makes it safe to read even with keys that aren't in the dictionary. If you try to read from the dictionary with a bad key, it should return 0 instead.

```
d = {1 : 2, 3 : 4}
safe_dict(d, 1) -> 2
safe_dict(d, 'cat') -> 0
```

````
def safe_dict(d,k):
    '''
    function safe_dict(d, k) that takes in a python dict d and a key 
    k and makes it safe to read even with keys that aren't in the dictionary.
    '''

    #keys(d) Returns a new object of the dictionary's keys.
    # if the key k is not in the dictionary-->False
    if (k in dict.keys(d))==False: 
        return 0
    else:
        return d[k]
d = {1 : 2, 3 : 4}
print(safe_dict(d, 1))
print(safe_dict(d, 'cat'))
````

# First NLP Program: IDF

Given a list of words, the the inverse document frequency (IDF) is a basic statistic of the amount of information of each word in the text.

The IDF formulat is:

$$IDF(w) = ln(\dfrac{N}{1 + n(w)})$$

Where:

- $w$ is the token (unique word),
- $n(w)$ is the number of documents that $w$ occurs in,
- $N$ is the total number of documents

Write a function, `idf(docs)` that takes in a list of lists of words and returns a dictionary  `word -> idf score`

Example:

```
IDF([['interview', 'questions'], ['interview', 'answers']]) -> {'questions': 0.0, 
                                                                'interview': -0.4, 
                                                                'answers': 0.0}


```

````
from pandas.core.common import flatten
import numpy as np 
def idf(doc):
    N=len(doc)

    #making a list of individual words in the doc
    a=list(set(flatten(doc)))
    wi=[]
    for word in a:
        w=0
        for i in range(N):
            
            if (word in doc[i])==True:
                w+=1
        wi.append(w)

    #list-->array
    wi=np.asanyarray(wi)

    #IDF formula
    IDF=np.log(N/(1+wi))


    dict1=dict(zip(a,IDF))
    return dict1
doc=[['interview', 'questions'], ['interview', 'answers']]

print(idf(doc))
````
